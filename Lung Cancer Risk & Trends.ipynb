```python
from scipy.stats import f_oneway
```

```python
# Run an ANOVA test to see if these different levels of pollution affect prevalence of lung cancer

f_statistic, p_value = f_oneway(low['Lung_Cancer_Prevalence_Rate'], medium['Lung_Cancer_Prevalence_Rate'], high['Lung_Cancer_Prevalence_Rate'])

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in the prevalence in lung cancer between different levels of air pollution exposure.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in the prevalence in lung cancer between different levels of air pollution exposure.")

```
